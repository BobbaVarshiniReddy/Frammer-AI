```python
import pandas as pd

df = pd.read_csv("channel-wise-publishing duration.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.replace("0:00:00", pd.NA)
```

```python
def convert_to_seconds(time_str):
    if pd.isna(time_str):
        return None
    h, m, s = map(int, time_str.split(':'))
    return h*3600 + m*60 + s

for col in df.columns[1:]:
    df[col] = df[col].apply(convert_to_seconds)
```

```python
df = df.fillna(0)
```

```python
df["Total Duration"] = df.iloc[:, 1:].sum(axis=1)
```

```python
df.head()
```

```python
import plotly.express as px

# Your color palette
custom_colors = [
    "#7A0000",
    "#CC0000",
    "#FF4649",
    "#FF9A9A",
    "#FFD6D6"
]

# Create correlation matrix
corr = df.iloc[:, 1:].corr()

# Create heatmap
fig = px.imshow(
    corr,
    text_auto=True,
    aspect="auto",
    color_continuous_scale=custom_colors,
    title="Correlation Heatmap"
)

# Apply dark theme + styling
fig.update_layout(
    template="plotly_dark",
    paper_bgcolor="black",
    plot_bgcolor="black",
    font=dict(color="white"),
    title_font=dict(size=20, color="white")
)

# Make numbers visible
fig.update_traces(
    textfont=dict(color="white")
)

# Show plot
fig.show()
```

```python
custom_colors = [
    "#7A0000",
    "#CC0000",
    "#FF4649",
    "#FF9A9A",
    "#FFD6D6"
]
platform_cols = df.columns[1:-1]

platform_sum = df[platform_cols].sum()
total_sum = platform_sum.sum()

platform_distribution = (platform_sum / total_sum) * 100

platform_distribution = platform_distribution.reset_index()
platform_distribution.columns = ["Platform", "Percentage"]

import plotly.express as px

fig = px.bar(
    platform_distribution,
    x="Platform",
    y="Percentage",
    title="Platform Publish Distribution (%)"
)

# Apply YOUR colors manually
fig.update_traces(
    marker=dict(
        color=platform_distribution["Percentage"],
        colorscale=custom_colors,
        showscale=True
    )
)

# Dark theme
fig.update_layout(
    template="plotly_dark",
    paper_bgcolor="black",
    plot_bgcolor="black",
    font=dict(color="white"),
    title_font=dict(color="white")
)

fig.show()
```

```python
zero_publish = df[(df.iloc[:, 1:-1] == 0).all(axis=1)]

zero_count = len(zero_publish)
active_count = len(df) - zero_count
```

```python
import plotly.express as px
import pandas as pd

kpi_df = pd.DataFrame({
    "Category": ["Zero Publish", "Active Channels"],
    "Count": [zero_count, active_count]
})

fig = px.bar(
    kpi_df,
    x="Category",
    y="Count",
    color="Category",
    color_discrete_sequence=[custom_colors[0], custom_colors[2]],
    title="Zero vs Active Channels"
)

fig.update_layout(
    template="plotly_dark",
    paper_bgcolor="black",
    plot_bgcolor="black",
    font=dict(color="white"),
    title_font=dict(color="white")
)

fig.show()
```
